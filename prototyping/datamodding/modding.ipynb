```python
import pandas as pd
DATA_PATH = "E:/VSProjekte/NutriGuard/prototyping/backend/data/dialysis.csv"
ID_START = 69420

df = pd.read_csv(DATA_PATH)
df
```

```python
def prepare_food_dataframe(df):
    # 1. Spalten umbenennen
    df = df.rename(columns={
        "foodname": "name",
        "kalium": "potassium"
    })
    
    df = df.sort_values('name', ascending=True)

    # 2. food_id generieren (z. B. fortlaufend)
    df["food_id"] = range(ID_START, len(df) + ID_START)

    # 3. Neue Spaltenreihenfolge definieren
    desired_order = [
        "food_id", "name", "category", 
        "kcal", "protein", "nacl", "potassium", "phosphate"
    ]

    # 4. DataFrame entsprechend umsortieren
    df = df[desired_order]
    df = df.reset_index(drop=True)

    return df

df = prepare_food_dataframe(df)
```

```python
df.to_csv(DATA_PATH, index=False)
```
